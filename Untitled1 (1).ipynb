# ***PEMBERSIHAN DATA (Clean Only)***

```python
# Import library
import pandas as pd
from google.colab import drive

# Mount Google Drive
drive.mount('/content/drive')

# Load dataset
df = pd.read_csv('/content/drive/MyDrive/dataset_estimasi_nilaimakanan/buahbuahan.csv')
df = pd.read_csv('/content/drive/MyDrive/dataset_estimasi_nilaimakanan/hewani.csv')
df = pd.read_csv('/content/drive/MyDrive/dataset_estimasi_nilaimakanan/nabati.csv')


# ====== Cek Missing Value ======
print("Jumlah missing value per kolom:")
print(df.isnull().sum())

# Hapus baris yang mengandung missing value
df = df.dropna()

# ====== Fungsi Hapus Outlier dengan IQR ======
def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

# Hapus outlier pada kolom 'kalori' dan 'protein'
df = remove_outliers(df, 'kalori')
df = remove_outliers(df, 'protein')

# Tampilkan bentuk data akhir
print("Data setelah dibersihkan:", df.shape)
```

# ***Preprocessing: encoding & pemisahan fitur-target***

```python
#Mengecek apakahh dataset terbaca atau tida (SEMENTTARA, NANTI DI HAPUS)

df_buah['kelas'] = 'buah'
df_nabati['kelas'] = 'nabati'
df_hewani['kelas'] = 'hewani'

df_all = pd.concat([df_buah, df_nabati, df_hewani], ignore_index=True)

print(df_all['kelas'].value_counts())
```

# ***Labeling fitur (X) dan target (y) ← LABELING BEFORE terjadi di sini***

```python
from sklearn.preprocessing import LabelEncoder

# Inisialisasi LabelEncoder
le_kelas = LabelEncoder()
le_kategori = LabelEncoder()

# Proses labeling (encoding)
df_all['kelas_encoded'] = le_kelas.fit_transform(df_all['kelas'])
df_all['kategori_encoded'] = le_kategori.fit_transform(df_all['kategori'])

# Cek hasilnya
print(df_all[['kelas', 'kelas_encoded']].drop_duplicates())
print(df_all[['kategori', 'kategori_encoded']].drop_duplicates())

df_all.head()
```

## ***Split data : Training Dan Testing*** (supervised learning)

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
import numpy as np
import pandas as pd

# Label Encoding satu kali saja
le_kategori = LabelEncoder()
df_all['kategori_encoded'] = le_kategori.fit_transform(df_all['kategori'])

le_kelas = LabelEncoder()
df_all['kelas_encoded'] = le_kelas.fit_transform(df_all['kelas'])

# Siapkan fitur dan target
X = df_all[['kategori_encoded', 'kelas_encoded', 'berat_gr']]
y_kalori = df_all['kalori'].values
y_protein = df_all['protein'].values

# Gabungkan target
y_combined = np.column_stack((y_kalori, y_protein))

# Split data
X_train, X_test, y_train_comb, y_test_comb = train_test_split(
    X, y_combined, test_size=0.2, random_state=42
)

# Pisahkan kembali
y_train_kal, y_train_pro = y_train_comb[:, 0], y_train_comb[:, 1]
y_test_kal, y_test_pro = y_test_comb[:, 0], y_test_comb[:, 1]

print("X_train sample:")
print(X_train.head())

print("y_train_kal sample:")
print(y_train_kal[:5])

print("y_train_pro sample:")
print(y_train_pro[:5])
```

# ***Pemilihan Fitur (Feature Selection)***

```python
from sklearn.ensemble import RandomForestRegressor
import pandas as pd

# Misal X dan y sudah siap
# X = DataFrame fitur
# y = target kalori atau protein

# Contoh training Random Forest
model_rf = RandomForestRegressor(random_state=42)
# Use y_kalori as the target for calculating feature importance
model_rf.fit(X, y_kalori)

# Ambil importance
importances = model_rf.feature_importances_

# Buat DataFrame untuk melihat ranking fitur
feature_importance_df = pd.DataFrame({
    'Fitur': X.columns,
    'Importance': importances
}).sort_values(by='Importance', ascending=False)

# Tampilkan
print(feature_importance_df)
```

# ***Tuning Hyperparameter untuk Kalori & Protein***

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor

# Parameter grid
param_grid = {
    'n_estimators': [100, 200],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2]
}

# GridSearchCV untuk model kalori
grid_kalori = GridSearchCV(
    estimator=RandomForestRegressor(random_state=42),
    param_grid=param_grid,
    cv=5,
    scoring='r2',
    n_jobs=-1,
    verbose=1
)
grid_kalori.fit(X_train, y_train_kal)
model_kalori_tuned = grid_kalori.best_estimator_
print("Best Params (Kalori):", grid_kalori.best_params_)

# GridSearchCV untuk model protein
grid_protein = GridSearchCV(
    estimator=RandomForestRegressor(random_state=42),
    param_grid=param_grid,
    cv=5,
    scoring='r2',
    n_jobs=-1,
    verbose=1
)
grid_protein.fit(X_train, y_train_pro)
model_protein_tuned = grid_protein.best_estimator_
print("Best Params (Protein):", grid_protein.best_params_)
```

# ***RandomizedSearchCV untuk kalori***

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import RandomizedSearchCV

# Parameter grid
param_dist = {
    'n_estimators': [100, 200, 300, 500],
    'max_depth': [None, 10, 20, 30, 50],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['sqrt', 'log2'] # Removed 'auto'
}

rf = RandomForestRegressor(random_state=42)

# RandomizedSearchCV untuk kalori
random_search_kal = RandomizedSearchCV(
    estimator=rf,
    param_distributions=param_dist,
    n_iter=30,
    cv=5,
    verbose=1,
    n_jobs=-1,
    scoring='neg_mean_squared_error',
    random_state=42
)

random_search_kal.fit(X_train, y_train_kal) # Changed X_train_kal to X_train
model_kal_best = random_search_kal.best_estimator_
print("Best Params Kalori:", random_search_kal.best_params_)
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# Load CSV dan beri label kategori
df_buah = pd.read_csv('/content/drive/MyDrive/dataset_estimasi_nilaimakanan/buahbuahan.csv')
df_buah['kategori'] = 'buahbuahan'

df_hewani = pd.read_csv('/content/drive/MyDrive/dataset_estimasi_nilaimakanan/hewani.csv')
df_hewani['kategori'] = 'hewani'

df_nabati = pd.read_csv('/content/drive/MyDrive/dataset_estimasi_nilaimakanan/nabati.csv')
df_nabati['kategori'] = 'nabati'

# Convert dan bersihkan masing-masing dataframe
for df in [df_buah, df_hewani, df_nabati]:
    df['berat_gr'] = pd.to_numeric(df['berat_gr'], errors='coerce')
    df['kalori'] = df['kalori'].astype(str).str.replace('.', '', regex=False)
    df['protein'] = df['protein'].astype(str).str.replace('.', '', regex=False)
    df['kalori'] = pd.to_numeric(df['kalori'], errors='coerce')
    df['protein'] = pd.to_numeric(df['protein'], errors='coerce')
    df.dropna(subset=['berat_gr', 'kalori', 'protein'], inplace=True)

# Gabungkan untuk proses Label Encoding kategori saja (bukan isi data)
gabungan_df = pd.concat([df_buah, df_hewani, df_nabati], ignore_index=True)

# Proses Label Encoding untuk kolom 'kategori'
le = LabelEncoder()
gabungan_df['kategori_encoded'] = le.fit_transform(gabungan_df['kategori'])

# Pisahkan lagi hasilnya ke 3 dataframe sesuai kategori
df_buah_after = gabungan_df[gabungan_df['kategori'] == 'buahbuahan']
df_hewani_after = gabungan_df[gabungan_df['kategori'] == 'hewani']
df_nabati_after = gabungan_df[gabungan_df['kategori'] == 'nabati']

# Tampilkan sebelum dan sesudah labeling
print("=== Dataset Buah-buahan Setelah Labeling ===")
display(df_buah_after)

print("=== Dataset Hewani Setelah Labeling ===")
display(df_hewani_after)

print("=== Dataset Nabati Setelah Labeling ===")
display(df_nabati_after)

# Menampilkan label yang dihasilkan
print("Label Mapping:")
print(dict(zip(le.classes_, le.transform(le.classes_))))
```
